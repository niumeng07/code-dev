```python
import pandas as pd
import requests

tkr = 'AMZN'
API_KEY = 'REU1BO1YXN'


url = f'https://www.alphavantage.co/query?function=EARNINGS&symbol={tkr}&apikey={API_KEY}'
proxy = 'http://127.0.0.1:8001'
proxies = None if proxy is None else {'http': proxy, 'https': proxy}

response = requests.get(url, proxies=proxies)

print(response.status_code)
```

```python
data = response.json()
```

```python
df = pd.DataFrame(data['quarterlyEarnings'])
```

```python
df
```

```python
df['Earnings Date'] = df['reportedDate']
df['EPS Estimate'] = df['estimatedEPS'].astype(float)
df['Reported EPS'] = df['reportedEPS'].astype(float)
df['Surprise(%)'] = df['surprisePercentage'].astype(float)/100
```
