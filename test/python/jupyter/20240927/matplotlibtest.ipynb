```python
import matplotlib.pyplot as plt
import pandas as pd

# 示例数据
data = pd.DataFrame({
    'Date': ['2023-01-01', '2023-02-01', '2023-03-01'],
    'A': [10, 20, 30],
    'B': [15, 25, 35]
})

# 将 'Date' 列转换为 pandas 日期时间格式（可选）
data['Date'] = pd.to_datetime(data['Date'])

# 创建图形和坐标轴
fig, ax = plt.subplots(figsize=(10, 6))

# 设置柱形宽度
bar_width = 0.35

# 设置x轴的位置
x = range(len(data))

# 绘制柱形图
bars1 = ax.bar([i - bar_width / 2 for i in x], data['A'], bar_width, label='A')
bars2 = ax.bar([i + bar_width / 2 for i in x], data['B'], bar_width, label='B')

# 设置x轴标签
ax.set_xticks(x)
ax.set_xticklabels(data['Date'].dt.strftime('%Y-%m-%d'), rotation=45)

# 添加标签和标题
ax.set_xlabel('Date')
ax.set_ylabel('Values')
ax.set_title('Bar Chart for A and B')
ax.legend()

# 显示图形
plt.tight_layout()
plt.show()
```
