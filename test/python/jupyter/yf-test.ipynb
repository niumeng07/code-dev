```python
import yfinance as yf
import pandas as pd
```

```python
ticker_data_file = '/Users/liuda/Local/data/trading/data/stocks/ticker/history_5m/AMZN.csv'
ticker_data = pd.read_csv(ticker_data_file, sep=",")
```

```python
ticker_data = ticker_data.assign(TICKER='AMZN')
```

```python
ticker_data
```

```python
ticker_data['Adj High'] = ticker_data['High'] + ticker_data['Adj Close'] - ticker_data['Close']
ticker_data['Adj Low'] = ticker_data['Low'] + ticker_data['Adj Close'] - ticker_data['Close']
ticker_data['Adj Open'] = ticker_data['Open'] + ticker_data['Adj Close'] - ticker_data['Close']
```

```python
ticker_data
```

```python
ticker_data = ticker_data.sort_values(by="Datetime").reset_index(drop=True)
```

```python
ticker_data['Datetime'] = pd.to_datetime(ticker_data['Datetime'])
ticker_data.set_index('Datetime', inplace=True)
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)
pd.set_option('display.width', 4000)

ticker_data.tail(5)
```

```python
resampled_data = ticker_data.resample(
	'D', label="left", offset="1D"
).agg({
	"Open": "first",
	"High": "max",
	"Low": "min",
	"Adj High": "max",
	"Adj Open": "first",
	"Adj Low": "min",
	"Adj Close": "last",
	"Volume": "sum",
})
```

```python
resampled_data.head(10)
```

```python
resampled_data.ffill().head(10)
```

```python
ticker_data = ticker_data.merge(resampled_data, on='Datetime', how='outer', suffixes=('', '_y'))
```

```python
import numpy as np

df = pd.DataFrame({'A': [1,2,3,4,5], 'B': [2,1,3,5,4]})
```

```python
output_col = 'test'

df[output_col] = 0
df['increment'] = 0
df.loc[np.where(df['A'] > df['B'], True, False), 'increment'] = 1
df.loc[np.where(df['A'] <= df['B'], True, False), 'increment'] = -1
                                           
df[output_col] = df['increment'].cumsum()
```
