```python
import pandas as pd
```

```python
df = pd.read_csv(f"/Users/liuda/Local/data/trading/data/KOL/YTB_KOL.csv")
SPY_trading_date = pd.read_csv(f'/Users/liuda/Local/data/trading/data/stocks/ticker/history/SPY.csv')
```

```python
SPY_trading_date
```

```python
df = df.drop(columns=['trading_date'])
```

```python
df['publish_date'] = pd.to_datetime(df['publish_date'], format='%Y-%m-%d', errors='coerce')
SPY_trading_date['Date'] = pd.to_datetime(SPY_trading_date['Date'], format='%Y-%m-%d', errors='coerce')
```

```python
df1 = pd.merge(df, SPY_trading_date, left_on='publish_date', right_on='Date', how='outer').reset_index(drop=True)
```

```python
df1['trading_date'] = df1['Date'].ffill()  # 线上使用
```

```python
# df[['publish_date']]
# SPY_trading_date[['Date']]
df1 = df1.dropna(subset=['publish_date'])
```
