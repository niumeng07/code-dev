```python
import mplfinance as mpf
import pandas as pd
import matplotlib.dates as mdates
import matplotlib.ticker as ticker
import matplotlib.pyplot as plt


# 示例数据
df = pd.read_csv('/Users/liuda/Local/data/finance/stocks/ticker/history/AMZN.csv', index_col=0, parse_dates=True)
max_volume = df['Volume'].max()
min_volume = df['Volume'].min()
df['Volume'] = df['Volume'] * 2 /(max_volume +min_volume )
print(df['Volume'])

# 生成一些示例数据
dates = pd.to_datetime(pd.date_range('2023-01-01', periods=100))

# 定义一个函数，用于在图表绘制后格式化 y 轴
def format_yaxis(ax):
    formatter = ticker.ScalarFormatter(useMathText=True)
    formatter.set_scientific(False) # 禁止科学计数法
    ax.yaxis.set_major_formatter(formatter)
    return ax

# 使用 mpf.plot 绘制图表
apds = [mpf.make_addplot(df['Volume'], panel=1, ylabel='Volume')]
fig, axes = mpf.plot(df,
        type='candle',
        style='yahoo',
        volume=True,
        addplot = apds, ylabel='', ylabel_lower='',
        returnfig=True)

# 应用自定义的 y 轴格式
for ax in axes:
    format_yaxis(ax)


plt.show()
```
